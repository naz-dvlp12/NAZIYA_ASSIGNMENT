Is gender independent of education level? A random sample of 395 people were surveyed and each person was asked to report the highest education level they obtained. The data that resulted from the survey is summarized in the following table: High School Bachelors Masters Ph.d. Total Female 60 54 46 41 201 Male 40 44 53 57 194 Total 100 98 99 98 395 Question: Are gender and education level dependent at 5% level of significance? In other words, given the data collected above, is there a relationship between the gender of an individual and the level of education that they have obtained?

```python
import numpy as np
from scipy import stats
import math
import pandas as pd
from scipy.stats import chi2_contingency

#H0: gender and qualification is independent
#H1: gender and qualification is not independent

dof = (4-1)*(2-1)

table = [[60, 54, 46, 41, 201] , [40, 44, 53, 57, 194]]
print(table)
print("\ndegrees of freedom:",dof)
chi2,p,dof,tbl = chi2_contingency(table)

print("\nChi2:",chi2)

print("\ntable_value:",tbl)

p_val = stats.chi2.pdf(8.006 , 3)

print("\np_value" , p_val)
print(p_val,"since the p value is less than the 5% level of significance , \nwe reject the null hypothesis, and conclude there is relationship between gender and the level of qualification tht they obtain.")
```

Using the following data, perform a oneway analysis of variance using Î±=.05. Write up the results in APA format. [Group1: 51, 45, 33, 45, 67] [Group2: 23, 43, 23, 43, 45] [Group3: 56, 76, 74, 87, 56]

```python
import numpy as np
from scipy import stats
import math
import pandas as pd

#H0: mu1 = mu2 = mu3
#H1: Atleast two means are different 

Group1 = [51, 45, 33, 45, 67]
Group2 = [23, 43, 23, 43, 45]
Group3 = [56, 76, 74, 87, 56]


df = pd.DataFrame(list(zip(Group1, Group2,Group3)), 
               columns =['Group1', 'Group2','Group3']) 
Group1_mean = np.mean(Group1)
print("\nGroup1_mean",Group1_mean,"Group2_mean:",Group2_mean,"Group3_mean:",Group3_mean)
Group2_mean = np.mean(Group2)
Group3_mean = np.mean(Group3)

df['G1-G1_mean'] = (df['Group1']-(Group1_mean))**2 
df['G2-G2_mean'] = (df['Group2']-(Group2_mean))**2
df['G3-G3_mean'] = (df['Group3']-(Group3_mean))**2

Grand_mean = np.mean(Group1+Group2+Group3)
print("Grand_mean",Grand_mean)

df['G1-Grand_mean'] = (df['Group1']-(Grand_mean))**2 
df['G2-Grand_mean'] = (df['Group2']-(Grand_mean))**2 
df['G3-Grand_mean'] = (df['Group2']-(Grand_mean))**2 

print(df)

SSE_1 = (df['G1-Grand_mean']).sum()
SSE_2 = (df['G2-Grand_mean']).sum()
SSE_3 = (df['G3-Grand_mean']).sum()
SST = SSE_1 + SSE_2 + SSE3
print("\nSSTotal:",SST)
SSE1 = (df['G1-G1_mean']).sum()
SSE2 = (df['G2-G2_mean']).sum()
SSE3 = (df['G3-G3_mean']).sum()
SSW = SSE1+SSE2+SSE3
print("\nSSwithin",SSW)

#Sum of square Between  = Sum of square total -sum of square within 
SSB = SST - SSW
print("\nSum of square between :", SSB)
print("\nDegree of freedom between samples:",(3-1))
print("\nDegree of freedom within samples:",(15-3))
MSb = SSB/2
print("\nMSb:",MSb)
MSw = SSW/12

print("\nMSw:",MSw)
F_cal = MSb/MSw
print("\nF calculated:", F_cal)
F_critical = stats.f.ppf(0.95,2,12)
print("\nF_critical:", F_critical)
print("\nCalculated value is greater than Critical value,\ntherefore we reject null hypothesis.\nand hence all means are not equal.")
```
